hospital 100

```python
hospital_100C = pd.read_csv('./data/datasets2/hospital_100_clean.csv')
hospital_100D = pd.read_csv('./data/datasets2/hospital_100_dirty.csv')
hosp_isClean = []
hosp_100C_txt = hospital_100C.to_string().splitlines()
hosp_100D_txt = hospital_100D.to_string().splitlines()

for i in range(1,len(hosp_100C_txt)):
    hosp_isClean.append(1 if ' '.join(hosp_100C_txt[i].split()) == ' '.join(hosp_100D_txt[i].split()) else 0)
    # print(' '.join(hosp_100C_txt[i].split()))
    # print(' '.join(hosp_100D_txt[i].split()))
    # print(hosp_100D_txt[i])
print(len(hosp_isClean))
```

```python
hospital_100D.insert(len(hospital_100D.columns),'isClean', hosp_isClean)
display(hospital_100D)
```

```python
hospital_100D.to_csv('./data/datasets2/hospital_100_dirty.csv', encoding='utf-8', index=False)
```
